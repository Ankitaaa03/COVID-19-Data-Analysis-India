```python
#Importing Libraries
import pandas as pd
import requests
import json
```

```python
#assigning url
timeseries_url = 'https://data.covid19india.org/v4/min/timeseries.min.json'
```

```python
# Making a GET request to the specified URL
r2 = requests.get(timeseries_url)
# Parsing the JSON response from the request
timeseries = r2.json()
```

```python
#converting into dataframe
df_timeseries = pd.DataFrame(timeseries)
```

```python
df_timeseries.head()
```

```python
#Verifying data across states
df_timeseries['AN'].loc['dates']
```

```python
df_timeseries.columns
```

## Creating DataFrame For each state

## AN

```python
# Create a DataFrame 'dfAN' by selecting the 'dates' column from the 'df_timeseries' DataFrame for the 'AN' index,
# transpose the result, and then convert each row into a separate column using apply(pd.Series)
dfAN = pd.DataFrame(df_timeseries['AN']['dates']).transpose().apply(pd.Series)
```

```python
dfAN
```

```python
# Drop the columns 'delta' and 'delta7' from the DataFrame 'dfAN' and apply the changes in-place
dfAN.drop(columns = ['delta','delta7'],inplace = True)
```

```python
# Extract and transform the 'total' column of the DataFrame 'dfAN' into a new DataFrame 'AN'
AN = dfAN['total'].apply(pd.Series)
```

```python
AN
```

```python
# Fill missing (NaN) values in the DataFrame 'AN' with zeros
AN = AN.fillna(0)
```

```python
# Drop the column named 'other' from the DataFrame 'AN'
AN = AN.drop(columns='other')
```

```python
AN
```

```python
# Reset the index of the DataFrame 'AN', keeping the current index as a regular column
AN.reset_index(level=0)
```

```python
AN.reset_index(level=0,inplace=True) # Reset the index of the DataFrame AN and apply the changes in-place
AN.rename({'index':'dates'},axis=1,inplace=True)# Rename the column 'index' to 'dates' in the DataFrame AN
AN['dates'] = pd.to_datetime(AN['dates']) # Reset the index of the DataFrame AN and apply the changes in-place
```

```python
# Convert the 'deceased' column in the DataFrame 'AN' to integers
AN['deceased'] = AN['deceased'].astype(int)
```

```python
AN.info()
```

```python
AN
```

```python
AN['deceased'].unique()
```
