```python
#Importing Libraries
import pandas as pd
import requests
import json
```

```python
#assigning url
data_url = 'https://data.covid19india.org/v4/min/data.min.json'
```

```python
# Making a GET request to the specified URL
r1 = requests.get(data_url)
# Parsing the JSON response from the request
data = r1.json()
```

```python
data
```

```python

# Initialize an empty list to store district data dictionaries
district = []

# Loop over all states in the JSON data
for state_code, state_data in data.items():
    # Loop over all districts within each state
    for district_code, district_data in state_data.get('districts', {}).items():
        # Extract relevant data categories for total counts
        confirmed = district_data.get('total', {}).get('confirmed', 0)
        recovered = district_data.get('total', {}).get('recovered', 0)
        deceased = district_data.get('total', {}).get('deceased', 0)
        tested = district_data.get('total', {}).get('tested', 0)
        vaccinated1 = district_data.get('total', {}).get('vaccinated1', 0)
        vaccinated2 = district_data.get('total', {}).get('vaccinated2', 0)
#        # Extract relevant data categories for delta changes
        confirmed_d = district_data.get('delta', {}).get('confirmed', 0)
        recovered_d = district_data.get('delta', {}).get('recovered', 0)
        deceased_d = district_data.get('delta', {}).get('deceased', 0)
        vaccinated1_d = district_data.get('delta', {}).get('vaccinated1', 0)
        vaccinated2_d = district_data.get('delta', {}).get('vaccinated2', 0)
#        # Extract relevant data categories for 7-day delta changes
        confirmed_d7 = district_data.get('delta7', {}).get('confirmed', 0)
        recovered_d7 = district_data.get('delta7', {}).get('recovered', 0)
        deceased_d7 = district_data.get('delta7', {}).get('deceased', 0)
        vaccinated1_d7 = district_data.get('delta7', {}).get('vaccinated1', 0)
        vaccinated2_d7 = district_data.get('delta7', {}).get('vaccinated2', 0)
        # Extract metadata information
        population_m = district_data.get('meta', {}).get('population', 0)
        tested_m = district_data.get('meta', {}).get('tested', {}).get('date', 0)
        last_updated_m = district_data.get('meta', {}).get('last_updated', 0)
        notes_m = district_data.get('meta', {}).get('notes', 0)
        date_m = district_data.get('meta', {}).get('date', 0)
        
#       # Create a dictionary for the current district's data
        district_dict = {
            'state_code': state_code,
            'district_code': district_code,
            #total
            'confirmed': confirmed,
            'recovered': recovered,
            'deceased': deceased,
            'tested': tested,
            'vaccinated1': vaccinated1,
            'vaccinated2': vaccinated2,
            #delta
            'confirmed_d': confirmed_d,
            'recovered_d': recovered_d,
            'deceased_d': deceased_d,
            'vaccinated1_d': vaccinated1_d,
            'vaccinated2_d': vaccinated2_d,
            #delta7
            'confirmed_d7': confirmed_d7,
            'recovered_d7': recovered_d7,
            'deceased_d7': deceased_d7,
            'vaccinated1_d7': vaccinated1_d7,
            'vaccinated2_d7': vaccinated2_d7,
            #meta
            'population_m': population_m,
            'tested_m' : tested_m,
            'last_updated_m' : last_updated_m,
            'notes_m' : notes_m,
            'date_m' : date_m
            
        }
        # Append the district dictionary to the list
        district.append(district_dict)

# Convert the list of dictionaries to a DataFrame
district_df = pd.DataFrame(district)
```
